```r
set.seed(1237)
n = 50000 # number of samples
numSong = 100 # number of song
prob = numeric(n)
y = numeric(n) # vector of sample that win
counter = 0
for (i in 1:n) {
    x = sample(1:numSong, numSong, repl = F)
    if (x[1] > 10 && x[2] > 10 && x[3] > 10 && x[4] > 10 &&
        x[5] >= 1 && x[5] <=10 ) {
            y[i] = 1
        counter = counter + 1
        }
    prob[i] = counter/i
}
# print(x)
# print(y)
p = mean(y == 1)
sprintf("P = %f", p)
plot(prob, ylab="Probability", xlab="Plot")
```
