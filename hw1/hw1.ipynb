# 2.

For this problem, as the question state that "any buffer is just as likely to ba available (or occupied) as any other", we can think that all buffers are same. So, we can think this as a combintation problem where order does not matter.

1. **A** = “A least two but no more than five buffers are occupied".  
We can think this **A** as the compliment of the possible that none of the buffers are occupied or only one buffer is occupied or all buffers are occupied. In this way, it is easier to do because each event is independent and we can simply just add them up.


**Solution:**

\begin{eqnarray*}
S = 2^6 = 64
\end{eqnarray*}

\begin{eqnarray*}
P(none occpuied) = \frac{1}{S} = \frac{1}{64} 
\end{eqnarray*}

\begin{eqnarray*} 
P(one occupied) = \frac{6}{S} = \frac{6}{64}
\end{eqnarray*}

\begin{eqnarray*} 
P(all occupied) = \frac{1}{S} = \frac{1}{64}
\end{eqnarray*}

\begin{eqnarray*} 
P(\overline{A}) = P(none occpuied) + P(one occupied) + P(all occupied) = \frac{8}{64}
\end{eqnarray*}

\begin{eqnarray*} 
P(A) = 1 - P(\overline{A}) = 1 - \frac{8}{64} = \frac{56}{64} = \frac{7}{8}
\end{eqnarray*}


2. B = “At least one buffer is occupied".  
Same as the question 1 which we can think B as the complement of the event that none buffer is
occupied.

**Solution**

\begin{eqnarray*}
    P(\overline{B}) = P(none occpuied) = \frac{1}{S} = \frac{1}{64}
    P(B) = 1 - P(\overline{B}) = 1 - \frac{1}{64} = \frac{63}{64}
\end{eqnarray*}

\begin{eqnarray*}
    P(B) = 1 - P(\overline{B}) = 1 - \frac{1}{64} = \frac{63}{64}
\end{eqnarray*}

**Solution**   
Set
event $A$ = the program will access the first location.  
event $B$ = the program will access the second location.  
event $(A \cap B)$ = the program will access both locations.  
event $(\overline{A \cup B})$ = the program will access neither location.

\begin{eqnarray*}
    P(A \cup B) = P(A) + P(B) - P(A \cap B) = 0.5 + 0.4 - 0.3 = 0.6
\end{eqnarray*}  

\begin{eqnarray*}
    P(\overline{A \cup B}) = 1 - P(A \cup B) = 1 - 0.6 = 0.4
\end{eqnarray*}
