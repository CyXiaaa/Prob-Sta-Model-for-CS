let's define two events.  
Event A: the first kid has green eyes.
Event B: the second kid hsa green eyes.
Since these two events are independent, so the probability that both of them have green eyes is the mutilply of the probability that one kid has green eyes.

\begin{eqnarray*}
P(A \cap B) = \frac{1}{2} * \frac{1}{2} = \frac{1}{4}
\end{eqnarray*}

using the same two events as in the part a). Now, given by the condition that the first kid is green eyes, find the probability that both two kids have green eyes. This is P(B|A).

\begin{eqnarray*}
P(B|A) = \frac{P(B \cap A)}{P(A)} = \frac{\frac{1}{4}}{\frac{1}{2}} = \frac{1}{2}
\end{eqnarray*}

```r
set.seed(1002)
numExp = 1000
numPeople = 1000
result = numeric(numPeople)
for(i in 1:numPeople){
    match = numeric(numExp)
    for(j in 1:numExp){
        x = sample(1:i, i, repl = F)
        for(k in 1:i){
            if(x[k] == k){
                match[j] = 1
                break;
            }
        }
    }
    
    p = mean(match == 1)
    result[i] = p
}

plot(result, ylab = "Probability", xlab="num of people")
abline(h = 1 - 1/exp(1), col = "red",lwd = 3)
```
