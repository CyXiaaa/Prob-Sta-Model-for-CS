```r
set.seed(1002)
numExp = 1000
numPeople = 1000
result = numeric(numPeople)
for(i in 1:numPeople){
    match = numeric(numExp)
    for(j in 1:numExp){
        x = sample(1:i, i, repl = F)
        for(k in 1:i){
            if(x[k] == k){
                match[j] = 1
                break;
            }
        }
    }
    
    p = mean(match == 1)
    result[i] = p
}

plot(result, ylab = "Probability", xlab="num of people")
abline(h = 1 - 1/exp(1), col = "red")
```
